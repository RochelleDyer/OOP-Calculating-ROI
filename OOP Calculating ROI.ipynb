```python
class ROI():
    #Make a dictionary for each category
    def __init__(self):
        self.dict = {
            'income': {
                'rental_income':0,
                'laundry':0,
                'storage':0,
                'misc':0,
                'total':0
            },
        
            'expenses': {
                'tax':0,
                'insurance':0,
                'utilities': {
                    'electric':0,
                    'water':0,
                    'sewer':0,
                    'garbage':0,
                    'gas':0,
                    },
                'hoa':0,
                'lawn_and_snow':0,
                'vacancy':0,
                'repairs':0,
                'capital_expenditures':0,
                'property_management':0,
                'mortgage':0,
                'total':0
            },
        
            'cash_flow': {
                'total_monthly_cash_flow' :0
            },
        
            'roi': {
                'down_payment':0,
                'closing_cost':0,
                'rehab_budget':0,
                'misc_other':0,
                'total_investment':0,
                'yearly_cash_flow':0,
                'cash_on_cash_roi':0
            }
        
        }
    
    def income(self):
        income_payments = []
        
        rental_income_payment = int (input("What are you charging for rent? "))
        income_payments.append (rental_income_payment)
        self.dict ['income']['rental_income'] = rental_income_payment
        
        laundry_income_payment = int(input('What are you charging for laundry? '))
        income_payments.append (laundry_income_payment)
        self.dict ['income']['laundry'] = laundry_income_payment
        
        storage_income_payment = int(input('What are you charging for storage? '))
        income_payments.append (storage_income_payment)
        self.dict ['income']['storage'] = storage_income_payment
        
        misc_income_payment = int(input('What are you charging for miscellaneous items? '))
        income_payments.append (misc_income_payment)
        self.dict ['income']['misc'] = misc_income_payment
        
        #Calculate the income total by adding all income expenses
        income_total = (self.dict ['income']['rental_income']) + (self.dict ['income']['laundry']) + (self.dict ['income']['storage'])
        + (self.dict ['income']['misc'])
        self.dict ['income']['total'] = income_total
        print(str(income_total) + " Total Income")
        
        
    def expenses(self):
        expense_payments = []
        
        tax_expense_payment = int(input('\nWhat do you pay in taxes per month? '))
        expense_payments.append (tax_expense_payment)
        self.dict ['expenses']['tax'] = tax_expense_payment
        
        insurance_expense_payment = int(input('What do you pay for insurance per month? '))
        expense_payments.append (insurance_expense_payment)
        self.dict ['expenses']['insurance'] = insurance_expense_payment
        
        electric_expense_payment = int(input('What do you pay for electricity per month? '))
        expense_payments.append (electric_expense_payment)
        self.dict ['expenses']['utilities']['electric'] = electric_expense_payment
        
        water_expense_payment = int(input('What do you pay for water per month? '))
        expense_payments.append (water_expense_payment)
        self.dict ['expenses']['utilities']['water'] = water_expense_payment
        
        sewer_expense_payment = int(input('What do you pay for sewage per month? '))
        expense_payments.append (sewer_expense_payment)
        self.dict ['expenses']['utilities']['sewer'] = sewer_expense_payment
        
        garbage_expense_payment = int(input('What do you pay for trash removal per month? '))
        expense_payments.append (garbage_expense_payment)
        self.dict ['expenses']['utilities']['garbage'] = garbage_expense_payment
        
        gas_expense_payment = int(input('What do you pay for gas per month? '))
        expense_payments.append (gas_expense_payment)
        self.dict ['expenses']['utilities']['gas'] = gas_expense_payment
        
        hoa_expense_payment = int(input('What do you pay for HOA fees per month? '))
        expense_payments.append (hoa_expense_payment)
        self.dict ['expenses']['hoa'] = hoa_expense_payment
        
        lawn_expense_payment = int(input('What do you pay for lawn care and snow removal per month? '))
        expense_payments.append (lawn_expense_payment)
        self.dict ['expenses']['lawn_and_snow'] = lawn_expense_payment
        
        vacancy_expense_payment = int(input('What do you pay for vacancy insurance per month? '))
        expense_payments.append (vacancy_expense_payment)
        self.dict ['expenses']['vacancy'] = vacancy_expense_payment
        
        repairs_expense_payment = int(input('What do you pay for repairs per month? '))
        expense_payments.append (repairs_expense_payment)
        self.dict ['expenses']['repairs'] = repairs_expense_payment
        
        capital_expenditure_payment = int(input('What do you pay for capital expenditures per month? '))
        expense_payments.append (capital_expenditure_payment)
        self.dict ['expenses']['capital_expenditures'] = capital_expenditure_payment
        
        property_management_payment = int(input('What do you pay for property management per month? '))
        expense_payments.append (property_management_payment)
        self.dict ['expenses']['property_management'] = property_management_payment
        
        mortgage_expense_payment = int(input('What do you pay for mortgage per month? '))
        expense_payments.append (mortgage_expense_payment)
        self.dict ['expenses']['mortgage'] = mortgage_expense_payment
        
        #Calculate total expenses
        expenses_total = sum (expense_payments)
        self.dict ['expenses']['total'] = expenses_total
        print(str(expenses_total) + " Total Expenses")
        
        
    def cash_flow(self):        
        income_total = self.dict['income']['total']
        print(str(income_total) + " Total Income")
        expenses_total = self.dict['expenses']['total']
        print(str(expenses_total) + " Total Expenses")
        total_cash_flow = income_total - expenses_total
        self.dict['cash_flow']['cash_flow_total'] = total_cash_flow
        print(str(total_cash_flow) + " Total Cash Flow")
        
    def cash_on_cash_roi (self):
        cash_on_cash_roi_payments = []
        
        down_payment = int(input('\nWhat was your down payment? '))
        cash_on_cash_roi_payments.append (down_payment)
        self.dict ['roi']['down_payment'] = down_payment
        
        closing_cost_payment = int(input('What were your closing costs? '))
        cash_on_cash_roi_payments.append (closing_cost_payment)
        self.dict ['roi']['closing_cost'] = closing_cost_payment
        
        rehab_budget_payment = int(input('How much did you spend on rehab for the property? '))
        cash_on_cash_roi_payments.append (rehab_budget_payment)
        self.dict ['roi']['rehab_budget'] = rehab_budget_payment
        
        misc_payment = int(input('What did you spend on miscellaneous items? '))
        cash_on_cash_roi_payments.append (misc_payment)
        self.dict ['roi']['misc_other'] = misc_payment
        
        #Calculate total investment by adding all items in cash on cash roi
        total_investment = (self.dict ['roi']['down_payment']) + (self.dict ['roi']['closing_cost']) + (self.dict ['roi']['rehab_budget']) + (self.dict ['roi']['misc_other'])
        self.dict['roi']['total_investment'] = total_investment
        print(str(total_investment) + " Total Investment")
        
        #Calculate yearly cash flow by taking monthly cash flow * 12
        yearly_cash_flow = (self.dict['cash_flow']['cash_flow_total']) * 12
        self.dict ['roi']['yearly_cash_flow'] = yearly_cash_flow
        print(str(yearly_cash_flow) + " Yearly Cash Flow")
        
        #Calculate cash on cash roi by taking yearly cash flow / total investment
        cash_on_cash_roi = (self.dict ['roi']['yearly_cash_flow']) / (self.dict['roi']['total_investment'])
        self.dict['roi']['cash_on_cash_roi'] = cash_on_cash_roi
        print(str((cash_on_cash_roi)*100) + " % Cash on Cash ROI")
        
    def run_program(self):
        self.income()
        self.expenses()
        self.cash_flow()
        self.cash_on_cash_roi() 

        
run = ROI()
run.run_program ()          
```
